## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df['subtotal'] = df['unit_price'] * df['qty']
print(df[['unit_price', 'qty', 'subtotal']].head())
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
df['shipping_price'] = (df['unit_weight'] * df['qty']).apply(lambda x: 7 * x if x > 50 else 10 * x)  
print(df[['unit_weight', 'qty', 'shipping_price']].head())
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
sales_tax = 0.0925
df['total_price'] = (df['subtotal'] + df['shipping_price']) * (1 + sales_tax)
print(df[['shipping_price', 'subtotal', 'total_price']].head())
```

```python
df.describe()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
df['line_cost'] = (df['unit_cost'] * df['qty']) * df['shipping_price']
print(df[['unit_cost', 'qty', 'shipping_price', 'line_cost']].head())
```

```python
# Create a column for the profit of each line using line cost and line price
df['line_profit'] = df['total_price'] - df['line_cost']
print(df[['line_cost', 'total_price', 'line_profit']].head())
```

You have email receipts showing that the total prices for 3 orders. Confirm that your calculations match the receipts. Remember, each order has multiple lines.

Order ID 2742071 had a total price of \$152,811.89

Order ID 2173913 had a total price of \$162,388.71

Order ID 6128929 had a total price of \$923,441.25


```python
order_ids = [2742071, 2173913, 6128929]
receipt_prices = [152811.89, 162388.71, 923441.25]
calculated_totals = df.groupby('order_id')['total_price'].sum().loc[order_ids].tolist()
order_totals = pd.DataFrame({
    'Order ID': order_ids,
    'Receipt Price': receipt_prices,
    'Calculated Price': calculated_totals
})
order_totals
```

```python
import numpy as np

tolerance = 1e-2
order_totals['Match'] = np.isclose(order_totals['Receipt Price'], order_totals['Calculated Price'], atol=tolerance)
for index, row in order_totals.iterrows():
    if row['Match']:
        print(f"Order ID {row['Order ID']} receipt matched the calculated total.")
    else:
        print(f"Order ID {row['Order ID']} receipt did not match the calculated total.")

order_totals
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
if isinstance(stored_top_5_clients, list):
    stored_top_5_clients = pd.DataFrame(stored_top_5_clients)
stored_top_5_clients['total_spent'] = df['unit_price'] * df['qty']
top_5_spending = stored_top_5_clients.groupby(df['client_id'])['total_spent'].sum().reset_index()
top_5_spending
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
stored_top_5_clients['total_units_purchased'] = df['unit_cost'] * df['qty']
stored_top_5_clients['total_revenue'] = df['unit_price'] * df['qty']
stored_top_5_clients['total_profit'] = (df['unit_price'] - df['unit_cost']) * df['qty']
stored_top_5_clients['total_shipping_price'] = df['unit_weight'] * df['qty']
summary_df = stored_top_5_clients.groupby(df['client_id']).agg(
    total_units_purchased=('total_units_purchased', 'sum'),
    total_shipping_price=('total_shipping_price', 'sum'),
    total_revenue=('total_revenue', 'sum'),
    total_profit=('total_profit', 'sum')
).reset_index()
summary_df
```

```python
# Format the data and rename the columns to names suitable for presentation.

# Define the money columns. 


# Define a function that converts a dollar amount to millions.


# Apply the currency_format_millions function to only the money columns. 


# Rename the columns to reflect the change in the money format. 
def to_millions(value):
    return value / 1_000_000
summary_df['total_shipping_price'] = summary_df['total_shipping_price'].apply(to_millions)
summary_df['total_revenue'] = summary_df['total_revenue'].apply(to_millions)
summary_df['total_profit'] = summary_df['total_profit'].apply(to_millions)
summary_df = summary_df.rename(columns={
    'client_id': 'Client ID',
    'total_units_purchased': 'Units',
    'total_shipping_price': 'Shipping (millions)',
    'total_revenue': 'Total Revenue (millions)',
    'total_profit': 'Total Profit (millions)'
})
summary_df

```

```python
summary_df['Shipping (millions)'] = summary_df['Shipping (millions)'] / 1e6
summary_df['Total Revenue (millions)'] = summary_df['Total Revenue (millions)'] / 1e6
summary_df['Total Profit (millions)'] = summary_df['Total Profit (millions)'] / 1e6
summary_df
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
sorted_summary_df = summary_df.sort_values(by='Total Profit (millions)', ascending=False)
sorted_summary_df
```
